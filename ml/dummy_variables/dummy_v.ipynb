```python
import pandas as pd
import numpy as np
import os
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
path = './'
read_csv = os.path.join(path, "house.csv")
df = pd.read_csv(read_csv , na_values = ['NA' , '?'])
```

```python
df
```

```python
df=df.drop(['view','waterfront','street','statezip','country','date','sqft_living','sqft_lot','floors','condition','sqft_above','sqft_basement','yr_renovated'],axis="columns")
```

```python
df
```

```python
for i in range(0,len(df)):
    b=df.yr_built[i]
    df.yr_built[i]=2020-b
```

```python
df1=pd.get_dummies(df.city)
```
