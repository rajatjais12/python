```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
%matplotlib inline
def gradent_desecnt(x,y):
    b_cunt=0
    m_cunt=0
    n=len(x)
    iterations=1000
    lr=0.01
    cost_previous = 0
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        y_predicted=m_cunt*x + b_cunt
        plt.plot(x,y_predicted,color='green')
        cost=-(1/n)*sum([val**2 for val in(y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_cunt=m_cunt-lr*md
        b_cunt=b_cunt-lr*yd
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        print("m {},b {},iteration {},cost {}".format(m_cunt,b_cunt,i,cost))
        
        
```

```python
x=np.array([1,2,4,6,7,8])
```

```python
y=np.array([12,18,23,28,34,39])
```

```python
 gradent_desecnt(x,y)
```
